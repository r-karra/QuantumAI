````python
import cirq
````

````python
# Pick a qubit
qubit = cirq.GridQubit(0, 0)

# Create a circuite that applies a square root of NOT gate, then measures the qubit.

circuit = cirq.Circuit(cirq.X(qubit) ** 0.5, cirq.measure(qubit, key='m'))
print("Circuite:")
print(circuit)

# Simulate the circuit several times
simulator = cirq.Simulator()
result = simulator.run(circuit, repetitions=20)
print("Results:")
print(result)
````

This code snippet uses the Cirq library to create and simulate a simple quantum circuit:

*   **`qubit = cirq.GridQubit(0, 0)`**: This line defines a single qubit at grid position (0, 0). In Cirq, qubits are often represented on a grid.
*   **`circuit = cirq.Circuit(cirq.X(qubit) ** 0.5, cirq.measure(qubit, key='m'))`**: This creates a quantum circuit.
    *   `cirq.X(qubit)` represents a Pauli-X gate (or NOT gate) applied to the `qubit`.
    *   `** 0.5` means applying the square root of the Pauli-X gate. This is a common operation in quantum computing.
    *   `cirq.measure(qubit, key='m')` measures the state of the `qubit` and stores the result with the key 'm'.
*   **`print("Circuite:")` and `print(circuit)`**: These lines print a textual representation of the quantum circuit.
*   **`simulator = cirq.Simulator()`**: This creates a simulator object, which allows you to run the quantum circuit on a classical computer.
*   **`result = simulator.run(circuit, repetitions=20)`**: This runs the defined `circuit` on the `simulator` 20 times. The `result` object contains the outcomes of these simulations.
*   **`print("Results:")` and `print(result)`**: These lines print the results of the simulation, showing the measurement outcomes from the 20 repetitions. In this case, the output shows a string of '0's and '1's, representing the measured state of the qubit in each repetition.

This code snippet uses `cirq.GridQubit.rect` to create a list of qubits arranged in a rectangular grid.

*   `cirq.GridQubit.rect(rows=3, cols=2)` creates a 3x2 grid of qubits. The qubits are indexed starting from (0, 0) and going up to (2, 1).
*   `print(myqubits)` then displays this list of qubits.

````python
myqubits = cirq.GridQubit.square(3)
print(myqubits)
````

This code snippet uses `cirq.GridQubit.square` to create a list of qubits arranged in a square grid.

*   `cirq.GridQubit.square(3)` creates a 3x3 grid of qubits. The qubits are indexed starting from (0, 0) and going up to (2, 2).
*   `print(myqubits)` then displays this list of qubits.

Imagine you have a box of building blocks, and each block is a tiny computer called a qubit!

*   `cirq.GridQubit.rect(rows=3, cols=2)` is like building a rectangular shape with your blocks. You decide how many rows of blocks you want (like 3) and how many blocks are in each row (like 2). It makes a shape that looks like this:


```
[block] [block]
[block] [block]
[block] [block]
```
* `cirq.GridQubit.square(3)` is like building a square shape with your blocks. You just say how many blocks you want on each side of the square (like 3). It makes a shape that looks like this:

```
[block] [block] [block]
[block] [block] [block]
[block] [block] [block]
```
So, `rect` is for making rectangles where the number of rows and columns can be different, and `square` is just a special kind of rectangle where the number of rows and columns are the same!

## Ways to create a qubit
* `cirq.NamedQubit()`
* `cirq.LineQubit(), cirq.LineQubit.range()`
* `cirq.GridQubit(), cirq.GridQubit.square(), cirq.GridQubit.rect()`

````python
myqubits = cirq.LineQubit.range(6)
mycircuit = cirq.Circuit()
mycircuit.append(cirq.H(myqubits[0]))
mycircuit.append(cirq.H(myqubits[1]))
mycircuit.append(cirq.H(myqubits[2]))
mycircuit.append(cirq.H(myqubits[3]))
mycircuit.append(cirq.H(myqubits[4]))
mycircuit.append(cirq.H(myqubits[5]))

print(mycircuit)
````

````python
# Create a circuit using loop

mycircuit = cirq.Circuit([cirq.H(qubit) for qubit in myqubits])
print(mycircuit)
````

````python
# Even more simpler than the above two methods

mycircuit = cirq.Circuit([cirq.H.on_each(myqubits)])
print(mycircuit)
````
